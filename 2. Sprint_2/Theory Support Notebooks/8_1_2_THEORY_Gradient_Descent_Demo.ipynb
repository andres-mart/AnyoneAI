# Gradient Descent Demo


````python
import matplotlib.pyplot as plt
import numpy as np
````

## Define Loss and Gradient

````python
def L(w):
    """
    Loss/Cost function.
    """
    return w ** 2 + 10


def grad_L(w):
    """
    Loss function Gradient.
    """
    return 2 * w


def tangent_line(w, xrange):
    """
    Tangent line at point w for Loss function.
    """
    return grad_L(w) * (xrange - w) + L(w)
````

## Visual representation

````python
weights = np.linspace(-10, 10, 50)

L_weights = L(weights)
grad_L_weights = grad_L(weights)

plt.figure(figsize=(7, 7))
plt.plot(weights, L_weights, "g", label="L(W)")
plt.plot(weights, grad_L_weights, "r.", label="L'(W)")
plt.legend(loc="best")
plt.show()
````

## Let's code Gradient Descent algorithm and plot it

````python
def gradient_descent_plot(w, lr=0.01):
    print(f"Initial weight value w = {w:.4f}")

    L_w = L(w)
    print(f"L(w) = {L_w:.4f}")

    grad_L_w = grad_L(w)
    print(f"L'(w) = {grad_L_w:.4f}")

    xrange = np.linspace(w-1, w+1, 10)
    w_tangent = tangent_line(w, xrange)

    # Gradient Descent step
    new_w = w - (lr * grad_L(w))
    new_loss = L(new_w)
    print(f"New weight is w' = {new_w:.4f} with Loss = {new_loss:.4f}\n")

    # Plot the figure
    weights = np.linspace(-10, 10, 50)
    L_weights = L(weights)
    plt.figure(figsize=(7, 7))
    plt.plot(weights, L_weights, "g", label="L(W)")
    plt.scatter(w, L_w, color='C1', s=50)
    plt.plot(xrange, w_tangent, 'C1--', linewidth = 2)
    plt.show()

    return new_w, new_loss
````

````python
_ = gradient_descent_plot(9, lr=0.01)
````

````python
weight = 9
prev_loss = 0
current_loss = float("+inf")

while abs(current_loss - prev_loss) >= 0.1:
    prev_loss = current_loss
    weight, current_loss = gradient_descent_plot(weight, lr=0.1)
````
