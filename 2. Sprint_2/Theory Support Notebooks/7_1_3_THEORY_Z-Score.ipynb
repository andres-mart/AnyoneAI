# Scaling data with Z-Score

$z = \frac{(x_i - u)}{s}$

Where `x` is a data sample, `u` is the mean value in our sample, and `s` is the standard deviation.

```python
from sklearn.preprocessing import StandardScaler
import pandas as pd
import warnings
warnings.filterwarnings('ignore')

data = pd.read_csv('data_escalado.csv')
data.head()
```

```python
data.describe()
```

```python
scaler = StandardScaler()
scaler.fit(data)
data_scaled = pd.DataFrame(scaler.transform(data), columns=['A','B','C'])
data_scaled.head()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(10,5))
sns.distplot(data.A, label='A');
sns.distplot(data.B, label='B');
sns.distplot(data.C, label='C');
plt.legend();
```

```python
plt.figure(figsize=(10,5))
sns.distplot(data_scaled.A, label='A');
sns.distplot(data_scaled.B, label='B');
sns.distplot(data_scaled.C, label='C');
plt.legend();
```
